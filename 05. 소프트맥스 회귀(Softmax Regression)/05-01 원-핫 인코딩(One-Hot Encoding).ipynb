#1. 원-핫 인코딩(One-hot encoding)이란?

원-핫 인코딩은 선택해야 하는 선택지의 개수만큼의 차원을 가지면서, 각 선택지의 인덱스에 해당하는 원소에는 1, 나머지 원소는 0의 값을 가지도록 하는 표현 방법이다.

#2. 원-핫 벡터의 무작위성

정수 인코딩은 {Banana :1, Tomato :2, Apple :3, Strawberry :4, ... Watermelon :10} 이런식으로 있을때 특정관계가 없음에도 loss function을 구했을때 바나나가 수박보다 토마토에 더 가깝다는 의미를 주게되는것 처럼 번호에 따라서 사용자가 부여하고자 했던 정보가 아닌게 부여될수도 있다.
